```python
from functools import partial
import matplotlib.pyplot as plt
import tensorflow as tf
Variable = tf.Variable
import tensorflow_probability as tfp
tfd = tfp.distributions
ed = tfp.edward2

print('TensorFlow version {}, TF Probability version {}.'.format(tf.__version__, tfp.__version__))
```

## Probabilistic PCA

Our underlying model is PPCA.
We have a dataset $X \in \mathbb{R}^{n \times d}$, which we aim to represent using low-dimensional latent variables $X \in \mathbb{R}^{n \times k}$.

We assume that $Z$ is Gaussian i.i.d.: 
$$ z_i \sim N(0, I), $$

and that each $x$ is generated by a shared projection matrix $W \in \mathbb{R}^{k \times d}$:
$$ x_i \sim N(Wz_i, \sigma^2 I). $$

```python
def ppca_model(d, k, n, std):
    w = ed.Normal(loc=tf.zeros([d, k]), scale=2.0*tf.ones([d, k]), name='w')
    z = ed.Normal(loc=tf.zeros([k, n]), scale=tf.ones([k, n]), name='z')
    x = ed.Normal(loc=tf.matmul(w, z), scale=std*tf.ones([d, n]), name='x')
    return x, (w, z)

log_joint = ed.make_log_joint_fn(ppca_model)
```

```python
d, k, n, std = 2, 1, 5000, .5
model = ppca_model(d=d, k=k, n=n, std=std)

with tf.Session() as sess:
    x_train, (_w, _z) = sess.run(model)

print('The true principal axis is {}'.format(_w))
plt.scatter(x_train[0], x_train[1], s=.1)
plt.arrow(0, 0, _w[0, 0], _w[1, 0], color='r', width=1e-1)
plt.show()
```

## MAP Inference

We now perform MAP inference, meaning we find the point estimate of $w, z$ that maximizes $p(w, z, x)$.

```python
w = Variable(tf.ones([d, k]))
z = Variable(tf.ones([k, n]))

target = partial(log_joint, d=d, k=k, n=n, std=std, x=x_train)
loss = -target(w=w, z=z)
optimizer = tf.train.AdamOptimizer(learning_rate=.1)
train = optimizer.minimize(loss)
init = tf.global_variables_initializer()

losses = []
with tf.Session() as sess:
    sess.run(init)
    for i in range(300):
        sess.run(train)
        if i % 10 == 0:
            losses.append(sess.run(loss))
    learned_w, learned_z = sess.run([w, z])
```

```python
print('Final loss was {}'.format(losses[-1]))
plt.plot(losses)
```

We can use [ed.interception](https://www.tensorflow.org/probability/api_docs/python/tfp/edward2/interception) to sample data for the our inferred values.

```python
def replace_lantents(w, z):
    def interceptor(rv_constructor, *rv_args, **rv_kwargs):
        name = rv_kwargs.pop('name')
        if name == 'w':
            rv_kwargs['value'] = w
        elif name == 'z':
            rv_kwargs['value'] = z
        return rv_constructor(*rv_args, **rv_kwargs)
    return interceptor

with ed.interception(replace_lantents(w=learned_w, z=learned_z)):
    generate = ppca_model(d=d, k=k, n=n, std=std)

with tf.Session() as sess:
    x_generated, _ = sess.run(generate)

plt.scatter(x_train[0], x_train[1], s=.2, label='train data')
plt.scatter(x_generated[0], x_generated[1], s=.2, c='r', label='generated data')
plt.legend()
```

## Variational Inference

Next, we use variational inference.
Instead of finding a point estimate of the parameters, we learn a parameterized distribution $q(w, z)$ and try to minimize the KL divergence between $q(\cdot)$ and the posterior $p(w, z | x)$.

```python
tf.reset_default_graph()

def variational_model(qw_mean, qw_stddv, qz_mean, qz_stddv):
    qw = ed.Normal(loc=qw_mean, scale=qw_stddv, name="qw")
    qz = ed.Normal(loc=qz_mean, scale=qz_stddv, name="qz")
    return qw, qz
variational_log_joint = ed.make_log_joint_fn(variational_model)

qw_mean = Variable(tf.zeros([d, k]))
qz_mean = Variable(tf.zeros([k, n]))
qw_stddv = tf.nn.softplus(Variable(-4 * tf.ones([d, k])))
qz_stddv = tf.nn.softplus(Variable(-4 * tf.ones([k, n])))
qw, qz = variational_model(qw_mean=qw_mean, qw_stddv=qw_stddv, qz_mean=qz_mean, qz_stddv=qz_stddv)

target_q = partial(variational_log_joint, qw_mean=qw_mean, qw_stddv=qw_stddv, qz_mean=qz_mean, qz_stddv=qz_stddv)
energy = target(w=qw, z=qz)
entropy = -target_q(qw=qw, qz=qz)
elbo = energy + entropy
optimizer = tf.train.AdamOptimizer(learning_rate=0.05)
train = optimizer.minimize(-elbo)
init = tf.global_variables_initializer()

elbos = []
with tf.Session() as sess:
    sess.run(init)
    for i in range(100):
        sess.run(train)
        if i % 5 == 0:
            elbos.append(sess.run([elbo]))
    learned_qw_mean, learned_qz_mean = sess.run([qw_mean, qz_mean])
```

```python
print('Final ELBO was {}'.format(elbos[-1]))
plt.plot(elbos)
```

```python
with ed.interception(replace_lantents(w=learned_qw_mean, z=learned_qz_mean)):
    generate = ppca_model(d=d, k=k, n=n, std=std)

with tf.Session() as sess:
    x_generated, _ = sess.run(generate)
plt.scatter(x_train[0], x_train[1], s=.2, label='train data')
plt.scatter(x_generated[0], x_generated[1], s=.2, c='r', label='generated data')
plt.legend()
```
