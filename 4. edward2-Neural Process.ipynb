```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_probability as tfp
tfd = tfp.distributions
psd_kernels = tfp.positive_semidefinite_kernels
ed = tfp.edward2
import warnings
warnings.filterwarnings('ignore')

print('TensorFlow version {}, TF Probability version {}.'.format(tf.__version__, tfp.__version__))
```

```python
# We define a 1d function regression task using samples from a Gaussian Process.
minx, maxx, numpoints, numtrain = -2., 2., 40, 20
xs, ys = [], []
for _ in range(16):
    x = tf.random_uniform((numpoints, 1), minval=minx, maxval=maxx)
    kernel = psd_kernels.ExponentiatedQuadratic(length_scale=1., amplitude=1)
    gp = tfd.GaussianProcess(kernel=kernel, index_points=x)
    xs.append(x)
    ys.append(4 * gp.sample())
task = (tf.stack(xs, 0), tf.expand_dims(tf.stack(ys, 0), 2))

init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
x, y = sess.run(task)
print("Let's look at some tasks sampled from our GP.")
for i in range(5):
    plt.scatter(x[i], y[i])
```

This implementation exploits the Helmholtz energy interpretation of the ELBO.

The ELBO can be rearranged as the sum of the negative energy and entropy of a particle determined by $z$,
where energy is defined as the joint probability $p(x, z)$, which edward2 can conveniently compute with ed.make_log_joint_fn().

```python
class Net():
    def __init__(self):
        super().__init__()
        self.xdim = 1
        self.zdim = 64
        self.hdim = 128
        self.global_step = tf.train.get_or_create_global_step()
        self.xytoz = self.subnet(self.hdim, self.zdim, 1, tf.nn.relu)
        self.zxtoy = self.subnet(self.hdim, self.xdim, 1, tf.nn.relu)
        self.log_joint = ed.make_log_joint_fn(self.generative_model)

    def subnet(self, hdim, outdim, hidden_n, activation):
        _network = tf.keras.Sequential()
        for _ in range(hidden_n):
            _network.add(tf.keras.layers.Dense(hdim, activation=activation))
        _network.add(tf.keras.layers.Dense(outdim * 2))
        return _network

    def generative_model(self, zloc, zdiag, x):
        z = ed.MultivariateNormalDiag(loc=zloc, scale_diag=zdiag, name='z')
        if len(z.shape) == 2:  # [bs, 2] to [bs, n, 2]. Is this the best way to do this?
            z = tf.tile(tf.expand_dims(z, 1), [1, x.shape[1].value, 1])
        zx = tf.concat([z, x], 2)
        outputs = self.zxtoy(zx)
        mu, _sigma = outputs[..., :self.xdim], outputs[..., self.xdim:]
        sigma = 0.1 + 0.9 * tf.nn.softplus(_sigma)
        y = ed.MultivariateNormalDiag(loc=mu, scale_diag=sigma, name='y')
        return x, y

    def variational_model(self, x, y, mode='dist'):
        assert mode in ['dist', 'params']
        xy = tf.concat([x, y], 2)
        outputs = self.xytoz(xy)
        outputs = tf.reduce_mean(outputs, axis=1)
        mu, _sigma = outputs[..., :self.zdim], outputs[..., self.zdim:]
        sigma = 0.1 + 0.9 * tf.nn.sigmoid(_sigma)
        z = ed.MultivariateNormalDiag(loc=mu, scale_diag=sigma, name='q_z')
        return (mu, sigma) if mode == 'params' else z

    def forward(self, data):
        full_x, full_y = data
        train_x, train_y = full_x[:, :numtrain], full_y[:, :numtrain]

        zloc, zdiag = self.variational_model(train_x, train_y, mode='params')
        q_full_z = self.variational_model(full_x, full_y)
        energy = self.log_joint(x=full_x, zloc=zloc, zdiag=zdiag, z=q_full_z, y=full_y)
        entropy = tf.reduce_sum(q_full_z.distribution.entropy())
        neg_elbo = -energy - entropy
        loss = neg_elbo / full_x.shape[0].value

        all_xs = tf.tile(tf.reshape(tf.range(minx-1, maxx+1, .01), [1, -1, 1]), [16, 1, 1])
        sampled_task = self.generative_model(zloc, zdiag, all_xs)
        return {'loss': loss, 'qz': q_full_z, 'data': data, 'sampled_task': sampled_task}
```

```python
model = Net()
model_outputs = model.forward(task)
opt = tf.contrib.opt.AdamWOptimizer(0, learning_rate=1e-3)
train_op = opt.minimize(model_outputs['loss'])
init = tf.global_variables_initializer()

sess = tf.Session()
sess.run(init)
for i in range(10001): # < 30 seconds on my gpu.
    _, _loss = sess.run([train_op, model_outputs['loss']])
    if i % 1000 == 0:
        print(i, _loss)
```

```python
qz, (data_x, data_y), (sampled_x, sampled_y) = sess.run(
        [model_outputs['qz'], model_outputs['data'], model_outputs['sampled_task']])
plt.figure(figsize=(16, 16))
for i in range(16):
    plt.subplot(4, 4, i + 1)
    plt.scatter(sampled_x[i, :, 0], sampled_y[i, :, 0], s=.2, label='generated')
    plt.scatter(data_x[i, :numtrain], data_y[i, :numtrain], label='train')
    plt.scatter(data_x[i, numtrain:], data_y[i, numtrain:], label='test')
plt.legend()
```
